```python
from bing_image_downloader import downloader
import pandas as pd

# Load CSV with no header
df = pd.read_csv("Boycott.csv", header=None)

# Set your custom column names
df.columns = ['product_name', 'category', 'description', 'country', 'manufacture']

# Display the DataFrame
df.head()
```

```python
# Filter for Egypt (country == 'EG')
egyptian_products = df[df['country'] == 'EG']['product_name'].unique()

# Download images for Egyptian products only
for name in egyptian_products:
    try:
        search_query = f"{name} Egypt"  # Append 'Egypt' to focus on local packaging
        downloader.download(search_query, limit=10, output_dir='product_images', adult_filter_off=True, force_replace=False)
    except Exception as e:
        print(f"Error downloading images for {name}: {e}")
```
